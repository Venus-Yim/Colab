```python
import pandas as pd

data = pd.read_csv('output.csv', header=None, names=['n_samples', 'my_pi', 'memory_required', 'elapsed_time', 'error'])

data.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

plt.figure(figsize=(8, 6))
plt.loglog(data['n_samples'], data['error'], 'bo-', markersize=4)
plt.xlabel('Number of Samples (N)', fontsize=12)
plt.ylabel('Relative Error', fontsize=12)
plt.title('N vs. Error', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.loglog(data['n_samples'], data['memory_required'], 'ro-', markersize=4)  # 
plt.xlabel('Number of Samples (N)', fontsize=12)
plt.ylabel('Memory Required (bytes)', fontsize=12)
plt.title('N vs. Memory Required', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.loglog(data['n_samples'], data['elapsed_time'], 'go-', markersize=4)  # 
plt.xlabel('Number of Samples (N)', fontsize=12)
plt.ylabel('Elapsed Time (seconds)', fontsize=12)
plt.title('N vs. Elapsed Time', fontsize=14)
plt.grid(True)
plt.show()
```
